# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
#Prob_3:Take another example and write the code that runs in linear amount of time no matter what is the size of input..
#similar to example-2, also plot the graph (eg. even_oddcounts elements greater than zero)

import time
import matplotlib.pyplot as plt
import random

def count_positives(data_list):

    count = 0
    for element in data_list:
        if element > 0:
            count += 1
    return count

def plot_time_complexity():

    input_sizes = []
    execution_times = []


    for size in range(1000, 51000, 2000):

        data_list = [random.randint(-100, 100) for _ in range(size)]

        start_time = time.time()
        count_positives(data_list)
        end_time = time.time()

        execution_time = end_time - start_time
        input_sizes.append(size)
        execution_times.append(execution_time)

    plt.figure(figsize=(10, 6))
    plt.plot(input_sizes, execution_times, marker='o', linestyle='-', color='b')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Time Complexity of count_positives() (Linear Time O(n))')
    plt.grid(True)
    plt.show()


if __name__ == '__main__':
    plot_time_complexity()

```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
import math
import matplotlib.pyplot as plt
def binary_search(arr,target):
    low = 0
    high = len(arr) - 1
    operations = 0
    while low <= high:
        operations += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return operations
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return operations

n_values = list(range(1, 100, 10))
time_values = []

for n in n_values:
    if n == 0:
        time_values.append(0)
        continue
    data = list(range(n))
    time_values.append(math.log2(n) if n > 0 else 0)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-', color='g')
plt.xlabel("Input size n")
plt.ylabel("Time (log2(n))")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search Approximation)")
plt.grid(True)
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.
import matplotlib.pyplot as plt
import time
import random
import math
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr

ns = [100, 500, 1000, 2000, 5000, 10000]
times = []

for n in ns:
    data = [random.randint(1, 10000) for _ in range(n)]
    start = time.time()
    merge_sort(data)
    end = time.time()
    times.append(end - start)

plt.figure(figsize=(10, 6))
plt.plot(ns, times, marker='o', linestyle='-', color='b', label='Merge Sort (O(n log n))')
plt.title('Log-Linear Time Complexity: Merge Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.grid(True)
plt.legend()
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
import time
import numpy as np
import matplotlib.pyplot as plt
import random

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]
    return merge(merge_sort(left_half), merge_sort(right_half))

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result
def run_and_measure_time(func, size):
    data = [random.randint(0, size) for _ in range(size)]
    start_time = time.perf_counter()
    func(data)
    end_time = time.perf_counter()
    return end_time - start_time


input_sizes = [10, 50, 100, 500, 1000, 2000, 5000, 10000, 15000, 20000]
execution_times = []


for size in input_sizes:
    time_taken = run_and_measure_time(merge_sort, size)
    execution_times.append(time_taken)


plt.figure(figsize=(10, 6))
plt.plot(input_sizes, execution_times, 'o-', label='Merge Sort Execution Time')


n_values = np.array(input_sizes)

k = execution_times[0] / (input_sizes[0] * np.log2(input_sizes[0]))
theoretical_times = k * n_values * np.log2(n_values)

plt.plot(input_sizes, theoretical_times, 'r--', label='O(n log n) Theoretical')

plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Merge Sort Time Complexity (O(n log n))")
plt.legend()
plt.grid(True)
plt.show()


```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example

import time
import matplotlib.pyplot as plt
import numpy as np
import random

def merge_sort(arr):
    """
    Sorts (O(n log n) time complexity).
    """
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left, right):

    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result
my_array = [4, 2, 3, 8, 8, 43, 6, 1, 0]
sorted_array = merge_sort(my_array)
print(f"Original array: {my_array}")
print(f"Sorted array: {sorted_array}")


def plot_merge_sort_performance():

    input_sizes = [10, 50, 100, 500, 1000, 2000, 5000, 10000]
    times = []

    for n in input_sizes:

        data = [random.randint(0, n) for _ in range(n)]
        start_time = time.perf_counter()
        merge_sort(data)
        end_time = time.perf_counter()
        times.append(end_time - start_time)

    plt.figure(figsize=(10, 6))
    plt.plot(input_sizes, times, 'o-', label='Merge Sort Execution Time')


    n_log_n_curve = [t / (n * np.log2(n)) if n > 1 else 0 for n, t in zip(input_sizes, times)]
    avg_scale = np.mean([x for x in n_log_n_curve if x > 0])
    theoretical = [avg_scale * n * np.log2(n) if n > 1 else 0 for n in input_sizes]
    plt.plot(input_sizes, theoretical, 'r--', label='Theoretical O(n log n)')

    plt.xlabel("Input Size (n)")
    plt.ylabel("Time Elapsed (seconds)")
    plt.title("Merge Sort Time Complexity: O(n log n)")
    plt.legend()
    plt.grid(True)
    plt.show()


plot_merge_sort_performance()
```

```python
# Write a function to perform quick sort and show that it has Logarithmic Time Complexity O(log n)
import time
import matplotlib.pyplot as plt
import numpy as np
import random

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quick_sort(left) + middle + quick_sort(right)
my_array = [4, 2, 3, 8, 8, 43, 6, 1, 0]
sorted_array = quick_sort(my_array)
print(f"Original array: {my_array}")
print(f"Sorted array: {sorted_array}")

def plot_quick_sort_performance():
    input_sizes = [10, 50, 100, 500, 1000, 2000, 5000, 10000]
    times = []

    for n in input_sizes:
        data = [random.randint(0, n) for _ in range(n)]
        start_time = time.perf_counter()
        quick_sort(data)
        end_time = time.perf_counter()
        times.append(end_time - start_time)

    plt.figure(figsize=(10, 6))
    plt.plot(input_sizes, times, 'o-', label='Quick Sort Execution Time')

    n_values = np.array(input_sizes)

    k_initial = times[0] / (n_values[0] * np.log2(n_values[0])) if n_values[0] > 1 else 0
    if np.isnan(k_initial) or np.isinf(k_initial):

        valid_k_points = [t / (n * np.log2(n)) for n, t in zip(n_values, times) if n > 1 and t > 0]
        k = np.mean(valid_k_points) if valid_k_points else 1e-7
    else:
        k = k_initial

    theoretical_times = k * n_values * np.log2(n_values)
    plt.plot(input_sizes, theoretical_times, 'r--', label='Theoretical O(n log n)')

    plt.xlabel("Input Size (n)")
    plt.ylabel("Time Elapsed (seconds)")
    plt.title("Quick Sort Time Complexity: O(n log n)")
    plt.legend()
    plt.grid(True)
    plt.show()


plot_quick_sort_performance()
```
